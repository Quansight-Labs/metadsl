```python
CHUNK_SIZE = 10

N_CHUNKS = 5

def create_chunk(chunk: int):
    return Vec.create_fn(
        Integer.from_int(CHUNK_SIZE),
        Abstraction[Integer, Integer].from_fn(lambda i: fetch_data(chunk, i))
    )

array = create_chunk(0)

for i in range(1, N_CHUNKS):
    array += create_chunk(i)
array = execute(HomoTupleCompat[IntCompat, Integer].from_maybe_vec(Maybe.just(array)))
```
